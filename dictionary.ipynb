```python
import os
import string
import nltk
import pandas as pd
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
def implement_stemmer(token):
    sno = nltk.stem.SnowballStemmer('english')
    return sno.stem(token)

stopwords = stopwords.words('english')
stopwords = [implement_stemmer(word) for word in stopwords]
stopwords
```

```python
df = pd.read_csv("paraphrases.csv")
df.head()
```

```python
NumberOfWords = 0
word_dict = {}

def tokenize_text(data):
    
    data = removePunctuation(data)
    data = data.lower()
    tokens = data.split()   
    return tokens

def removePunctuation(text):
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text

check = df.sentence1 + df.sentence2
for sentence in check:
    stemmed_list = []
    tokens = tokenize_text(sentence)
    
    for word in tokens:
        stem_word = implement_stemmer(word)
        if stem_word not in stopwords:
            stemmed_list.append(stem_word)
    NumberOfWords+= len(stemmed_list)
    for i in stemmed_list:
        if i in word_dict:
            word_dict[i]+=1
        else:
            word_dict[i] = 1
word_dict
```

```python
d = word_dict
topwords = []
for w in sorted(d, key=d.get, reverse=True):
    topwords.append([w, d[w]])
print("Word Frequency")
for i in range(100):
    print(topwords[i][0],topwords[i][1])
```
